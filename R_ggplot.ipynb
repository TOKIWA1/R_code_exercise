使用分面的优势：变量的每个参数独立且清楚，适用于参数之间无比较的图
使用图形属性的优势：变量的每个参数可相互比较，但图形较大且复杂，容易压缩图像

```r
?facet_wrap
```

nrow:每一行的个数；ncol:每一列的个数，facet_grid()没有nrow和ncol是因为它的row和col被row中的变量和col中的变量所代替，是固定值，其分组也并非单一分组，而是根据两个变量进行分组。

布局好看，横向分类过多容易压缩图形，所以纵向分类。

## 1.7 统计变换

```r
ggplot(data=diamonds)+
geom_bar(mapping=aes(x=cut))
```

条形图x轴显示的是cut，y轴显示的是count，count来自计算出的新数据

条形图、直方图和频率图对数据进行分类，绘制分类和落在每个分类的数据点数量；
平滑曲线会拟合模型，然后绘制拟合模型预测值；
箱线图计算数据分布的多种摘要统计量；

```r
ggplot(data=diamonds)+
stat_count(mapping=aes(x=cut))
```

显式使用某统计变量的三种原因

```r
demo <- tribble(
    ~a,~b,
    "bar_1",20,
    "bar_2",30,
    "bar_3",40)
```

```r
demo
```

```r
ggplot(data=demo)+
geom_bar(
mapping=aes(x=a,y=b), stat="identity")
```

2.覆盖从统计变换生成的变量到图形属性的默认变换

```r
ggplot(data=diamonds)+
geom_bar(mapping=aes(x=cut,y=..prop..,group=1))
```
