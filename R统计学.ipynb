建立线性回归，制作表格

```r
# 建立线性回归模型
linear_model <- lm(mpg ~ am + drat + wt + hp, data = mtcars)

# 制作回归结果的表格
linear_model %>% tbl_regression()
```

```r
linear_model %>% tbl_regression() %>% add_inline_forest_plot()
```
